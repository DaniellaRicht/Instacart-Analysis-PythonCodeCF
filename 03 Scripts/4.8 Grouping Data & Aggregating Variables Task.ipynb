## 0.4 Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Creating new column "max_order"

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check for column

ords_prods_merge.head(15)
```

```python
# Create a flag that assigns loyalty label to user id based on corresponding max order value

ords_prods_merge.loc[ords_prods_merge['max_order']>40,'loyalty_flag']='Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order']<=40) & (ords_prods_merge['max_order']>10),'loyalty_flag']='Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<=10,'loyalty_flag']='New customer'
```

```python
# Check loyalty flag value count

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check Loyalty flag column in dataframe

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(15)
```

## 0.5 Use the loyalty flag & check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer) to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers

```python
# Check the min, max, mean of loyalty flag

ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

#### Looking at the above numbers, we can determine that New customers are (by mean) spending the most, whilst Loyal customers are spending the least. 

## 0.6 Create a spending flag for each user based on the average price across all their orders using the following criteria

```python
# Create a column for average spender

ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create a “spending flag,” for high and low spender

ords_prods_merge.loc[ords_prods_merge['average_spend']<10,'spender_flag']='Low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check the Spender flags value count

ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Check Spender flag column in dataframe

ords_prods_merge[['user_id', 'average_spend', 'spender_flag', 'prices']].head(20)
```

## 0.7 Determine frequent versus non-frequent customers

```python
# Create a new column Customer Frequency based on the median in the “days_since_prior_order” column

ords_prods_merge['Customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check for new column

ords_prods_merge.head(15)
```

```python
# Create "Customer Frequency" flag on order frequency based on “days_since_prior_order”

ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['Customer_frequency']>10) & (ords_prods_merge['Customer_frequency']<=20),'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check frequency flags value counts

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Look at the newly added flag cloumn

ords_prods_merge[['user_id','Customer_frequency','frequency_flag']].head(50)
```
