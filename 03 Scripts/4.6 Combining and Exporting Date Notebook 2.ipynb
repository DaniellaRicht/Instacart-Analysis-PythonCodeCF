## 3. Determine a suitable way to combine the orders_products_combined dataframe with your products data set.

```python
# Drop unnecessary columns in df prods

df_prods = df_prods.drop(['Unnamed: 0'], axis=1)
```

```python
# Check the shape

df_prods.shape
```

#### I believe a suitable way to combine both dataframes would be using their shared column 'product_id', using a default inner join.

```python
# Merge df_prods and df_ords_combined

df_merged = df_prods.merge(df_ords_prods_combined, on = 'product_id', indicator = True)
```

#### I can see from the ValueError above that the name of an exisiting column is used for the indicator column and this doesn't work. I believe this is because df_ords_prods_combine has already been merged once and therefore has a _merge column.

```python
# Drop unnecessary columns
df_ords_prods_combined = df_ords_prods_combined.drop(['_merge'], axis=1)
```

```python
# Check output 
df_ords_prods_combined.head()
```

```python
# Merging the updated df_ords_prods_combined 

df_ords_prods_merge = df_prods.merge(df_ords_prods_combined, on = 'product_id', indicator = True)
```
