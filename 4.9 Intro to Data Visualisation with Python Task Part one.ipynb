## 0.4 Combine Data

```python
# Task 6 Combine your customer data with the rest of your prepared Instacart data
# Import dataset orders_products_merge_agg.pkl

df_ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data ', 'Prepared Data', 'ords_prods_merge_agg.pkl'))
```

```python
# Check ords_prods_merge_agg for key columns

df_ords_prods_merge.head()
```

```python
#Drop the _merge column from the df_ords_prods_merge dataframe as it is not needed

df_ords_prods_merge = df_ords_prods_merge.drop(columns = ['_merge'])
```

```python
# Check ords_prods_merge_agg for data types

df_ords_prods_merge.dtypes
```

```python
#Change the data type of user_id to string

df_ords_prods_merge['user_id'] = df_ords_prods_merge['user_id'].astype('int64')
```

```python
# Check for changes

df_ords_prods_merge.info()
```

```python
# Merge df_ords_prods_merge with df_customers using the user_id key


df_ords_prods_cust = df_ords_prods_merge.merge(df_customers, on=['user_id'], indicator=True)
```
